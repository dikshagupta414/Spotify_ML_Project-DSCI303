```python
import pandas as pd
import numpy as np
import sklearn as sklearn
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
from sklearn.manifold import TSNE
```

```python
df = pd.read_csv('~/Box/DSCI303/archive/final_data.csv')
df.head()
```

```python
features = ['acousticness', 'danceability', 'energy', 'instrumentalness',
        'liveness', 'loudness','speechiness','tempo','valence']
df_kmeans = df[features]
```

```python
### PCA with N Components = 2
### Also performed K Means

pca = PCA(n_components = 2)
X = pca.fit_transform(df_kmeans)

X_new = pd.DataFrame(X)

X_new['year_bins'] = df['year_bins']

year0 = X_new[X_new['year_bins'] == 0]
year1 = X_new[X_new['year_bins'] == 1]
year2 = X_new[X_new['year_bins'] == 2]
year3 = X_new[X_new['year_bins'] == 3]
year4 = X_new[X_new['year_bins'] == 4]

plt.scatter(year0[0], year0[1], color = 'red')
plt.scatter(year1[0], year1[1], color = 'orange')
plt.scatter(year2[0], year2[1], color = 'yellow')
plt.scatter(year3[0], year3[1], color = 'black')
plt.scatter(year4[0], year4[1], color = 'blue')
```

```python
X_new = pca.inverse_transform(X)
kmeans = KMeans(n_clusters=5).fit(X)

X_new = pca.transform(X_new)
X_new = pd.DataFrame(X_new)
X_new['labels'] = kmeans.labels_
X_new['year_bins'] = df['year_bins']

year0 = X_new[X_new['year_bins'] == 0]
year1 = X_new[X_new['year_bins'] == 1]
year2 = X_new[X_new['year_bins'] == 2]
year3 = X_new[X_new['year_bins'] == 3]
year4 = X_new[X_new['year_bins'] == 4]
```
