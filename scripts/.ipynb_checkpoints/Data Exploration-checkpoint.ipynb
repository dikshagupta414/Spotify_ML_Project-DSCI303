 # Data Exploration and Wrangling
 We use scatter plots, correlation plots, PCA and various other techniques for data visualization and exploration. 

```python
import pandas as pd
import numpy as np
import sklearn as sklearn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../archive/data.csv')
df_w_genres = pd.read_csv('../archive/data_w_genres.csv')
dfa = pd.read_csv('../archive/data_by_artist.csv')
dfy = pd.read_csv('../archive/data_by_year.csv')
dfg = pd.read_csv('../archive/data_by_genres.csv')
# df.info()
# df_w_genres.info()
# dfg.info()
print(df.shape)
# print(df_w_genres.shape)
# print(dfg.shape)
# df_w_genres.head()
```

```python
# Data Wrangling
# Create a new dataset called "final_data.csv" to include new features we created as well.

## Normalize loudness and tempo_norm - 'loudness_norm' and 'tempo_norm'
df['loudness_norm'] = (df['loudness'] - np.mean(df['loudness'])) / np.std(df['loudness'])
df['temp_norm'] = (df['tempo'] - np.mean(df['tempo'])) / np.std(df['tempo'])


## Create bins for years - 'year_bins'
df['year_bins'] = pd.cut(df['year'], bins=[1920, 1940, 1960, 1980, 2000, 2020], labels=range(5))

# print(df['year_bins'])
# print(pd.cut(df['year'], bins=[1920, 1940, 1960, 1980, 2000, 2020]).value_counts())


## Create bins for popularity 
### "popularity_cut" - equal ranges. - 4 bins
### "popularity_cut_labels" - labeled popularity.
### "popularity_qcut" - Quantile-based cuts = equal sized bins - 4 bins
### "popularity_qcut_labels"
df['popularity_cut'] = pd.cut(df['popularity'], bins=4, labels=range(4))
df['popularity_cut_labels'] = pd.cut(df['popularity'], bins=4)#labels=['most unpopular','somewhat unpopular', 'somewhat popular','most popular']
print(pd.cut(df['popularity'], bins=4, labels=range(4))) #'most unpopular' -> 'somewhat unpopular' -> 'somewhat popular' -> 'most popular'
print(pd.cut(df['popularity'], bins=4).value_counts())

df['popularity_qcut'] = pd.qcut(df['popularity'], q=4, labels=range(4))
df['popularity_qcut_labels'] = pd.cut(df['popularity'], bins=4) #labels=['most unpopular','somewhat unpopular', 'somewhat popular','most popular'])
print(pd.qcut(df['popularity'], q=4, labels=range(4))) #'most unpopular' -> 'somewhat unpopular' -> 'somewhat popular' -> 'most popular'
print(pd.qcut(df['popularity'], q=4).value_counts())

df.to_csv('../archive/final_data.csv')
```

## Correlation Matrix
We found that there is a significant correlation between:
1. positive correlation between loudness and energy 
2. positive correlation between valence and danceability 
3. negative correlation loudness/energy and acoustiveness 
4. positive correlation between year and popularity

```python
#Entire dataset
plt.figure(figsize=(16, 8))
plt.title("Correlation for data")
sns.set(style="whitegrid")
corr = df.corr()
sns.heatmap(corr,annot=True)

#Dataset with genres included
plt.figure(figsize=(16, 8))
plt.title("Correlation for data WITH genre")
sns.set(style="whitegrid")
corr = df_w_genres.corr()
sns.heatmap(corr,annot=True)

#Dataset by artist
plt.figure(figsize=(16, 8))
plt.title("Correlation for data by artist")
sns.set(style="whitegrid")
corr = dfa.corr()
sns.heatmap(corr,annot=True)

#Dataset by year
plt.figure(figsize=(16, 8))
plt.title("Correlation for data by year")
sns.set(style="whitegrid")
corr = dfy.corr()
sns.heatmap(corr,annot=True)

#Dataset by genre
plt.figure(figsize=(16, 8))
plt.title("Correlation for data by genre")
sns.set(style="whitegrid")
corr = dfg.corr()
sns.heatmap(corr,annot=True)
```

## Scatter Plots
To further explore the correlation we plotted the scatter plots for the correlated variable.

```python
# Create a figure and axis
fig, ax = plt.subplots(4, figsize=(10,20))
x_list = ['year','loudness', 'valence', 'energy']
y_list = ['popularity','energy', 'danceability', 'acousticness']
for i in range(len(x_list)):
    x=x_list[i]
    y=y_list[i]
    # Scatter the x against the y
    ax[i].scatter(df[x], df[y],c=df['popularity'], s=0.1)
    # set a title and labels
    ax[i].set_title('Spotify Dataset')
    ax[i].set_xlabel(x)
    ax[i].set_ylabel(y)
```

## Time Series Analysis

```python
plt.figure(figsize=(16, 10))
sns.set(style="whitegrid")
columns = ["acousticness","danceability","energy","speechiness","liveness","valence", "loudness_norm"] #"loudness"
for col in columns:
    x = df.groupby("year")[col].mean()
    ax= sns.lineplot(x=x.index,y=x,label=col)
ax.set_title('Normalized audio characteristics over year')
ax.set_ylabel('Measure')
ax.set_xlabel('Year')
```
