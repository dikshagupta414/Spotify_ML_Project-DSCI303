```python
#t-SNE
import pandas as pd
import numpy as np
import sklearn as sklearn
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
from sklearn.manifold import TSNE
import time
```

```python
df = pd.read_csv('~/Box/DSCI303/archive/final_data.csv')
df_w_genres = pd.read_csv('~/Box/DSCI303/archive/data_w_genres.csv')
dfa = pd.read_csv('~/Box/DSCI303/archive/data_by_artist.csv')
dfy = pd.read_csv('~/Box/DSCI303/archive/data_by_year.csv')
dfg = pd.read_csv('~/Box/DSCI303/archive/data_by_genres.csv')
df.info()
df.head()
```

```python
df_test = pd.read_csv('~/Box/DSCI303/archive/final_year_test.csv')
df_train = pd.read_csv('~/Box/DSCI303/archive/final_year_train.csv')

features = ['acousticness', 'danceability', 'energy', 'instrumentalness',
            'liveness', 'loudness','speechiness','tempo','valence']
```

```python
# perplexities =[10, 50, 100, 500,1000,1500]
perplexities=[500]
#t-SNE
for i,perp in enumerate(perplexities):
    time_start = time.time()
    tsne = TSNE(n_components=2, verbose=2, perplexity=perp, n_iter=250)
    tsne_results = tsne.fit_transform(df[features].values)
    print('t-SNE done! Time elapsed: {} seconds'.format(time.time()-time_start))
    df['tsne-2d-one'] = tsne_results[:,0]
    df['tsne-2d-two'] = tsne_results[:,1]
    plt.figure(figsize=(16,10))
    plt.title('tSNE perp='+str(perp))
    sns.scatterplot(
        x="tsne-2d-one", y="tsne-2d-two",
        hue="year_bins",
        palette=sns.color_palette("hls", 4),
        data=df,
        legend="full",
        alpha=0.3)
    plt.show()
#     plt.savefig('tSNE_'+str(perp)+'.png')
```
